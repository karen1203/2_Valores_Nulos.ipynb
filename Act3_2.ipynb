```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Factura=pd.read_excel('Factura.xlsx')
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```

En CVE_DOC logré identificar que es el id de factura por lo que no se ñpuede llenar con algún promedio, por lo que si está vacío algún espacio en esa columna es mejor dejarlo como un documento desconocido.

```python
Factura["CVE_DOC"]=Factura["CVE_DOC"].fillna('Desconocido')
Factura
```

```python
df6=Factura.groupby(['CVE_VEND'])['CVE_VEND'].count()
df6
```

En la variable de CVE_VEND dado que solo hay dos posibles respuestas lo remplazaré con la más frecuente, es decir por la moda.

```python
Factura["CVE_VEND"]=Factura["CVE_VEND"].fillna(7706)
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```

```python
df7=Factura.groupby(['STATUS'])['STATUS'].count()
df7
```

Al igual que en la variable anterior, en STATUS remplazamos los valores faltantes por los más repetidos que en este caso es E

```python
Factura["STATUS"]=Factura["STATUS"].fillna('E')
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```

En todas las fechas se reemplazará con los datos anteriores, pues las facturas deberían de estar organizadas por fechas consecutivas, por lo que es posible que sean de la misma fecha que la anterior, no hay mucha varianza.

```python
Factura["FECHAELAB"]=Factura["FECHAELAB"].fillna(method='bfill')
Factura
```

```python
Factura["FECHA_ENT"]=Factura["FECHA_ENT"].fillna(method='bfill')
Factura
```

```python
Factura["FECHA_DOC"]=Factura["FECHA_DOC"].fillna(method='bfill')
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```

En la variable de fechas de cancelación decidí no borrar la columna y agregar un texto que diga que no se ha cancelado, pues es importante mantener el registro de cada factura auqnue se haya cancelado, ya que si se realiza en algún momento alguna auditoría también podría necesitarse ese registro.

```python
Factura["FECHA_CANCELA"]=Factura["FECHA_CANCELA"].fillna('NO ha sido cancelada')
Factura
```

Para remplazar los valores nulos en CAN_TOT decidí usar el promedio de la columna ya que son valores numéricos que pueden alejarse un poco entre ellos.

```python
Factura["CAN_TOT"]=Factura["CAN_TOT"].fillna(round(Factura["CAN_TOT"].mean(),1))
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```

```python
df8=Factura.groupby(['DES_TOT'])['DES_TOT'].count()
df8
```

En la variable DES_TOT se va a llenar con 0.00 ya que la mayoría de los valores de esa columna son 0


```python
Factura["DES_TOT"]=Factura["DES_TOT"].fillna(0.00)
Factura
```

```python
df9=Factura.groupby(['DES_FIN'])['DES_FIN'].count()
df9
```

En DES_FIN lo cambié por 0 también pues todos los demás valores de la columna tienen 0.

```python
Factura["DES_FIN"]=Factura["DES_FIN"].fillna(0.00)
Factura
```

```python
nulos=Factura.isnull().sum()
nulos
```
